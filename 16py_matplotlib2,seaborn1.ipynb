adult_female column을 만들고 나이(age)가 20이상이고 성별(sex)이 'female'인 값은  
True를 아니면 False 값을 할당합니다.

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic = titanic[titanic.age.notnull()]
titanic['adult_female'] = titanic.apply(lambda x: 'True' if x.age >= 20 and x.sex == 'female' else 'False', axis = 1)
titanic
```

```python
import numpy as np

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

## 컨투어 플롯

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title('Contour plots')
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

## 3D 서비스 플롯

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('3D Surface Plot')
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```

## 히스토그램 - histplot() 다양한 표현법 poly

```python
# When both x and y are assigned,
# a bivariate histogram is computed and shown as a heatmap
sns.histplot(penguins, x='bill_depth_mm', y='body_mass_g', hue='species')
```
